```python
import numpy as np
import math
from scipy import interpolate
from scipy.stats import linregress
import matplotlib.pyplot as plt
import matplotlib
import bspline
import bspline.splinelab
```

```python
splev_res = 1000

def softmax(w):
    w=np.exp(w)
    w/=w.sum()
    return w

def gdk(z, n, k):
    zt = z[1:-1]
    t = []
    for i in range(n):
        t.append(zt[round((len(zt)-1)*(i+1)/(n+1))])
        
    t = np.concatenate(([z.min()]*(k+1), t, [z.max()]*(k+1)))
    return t

def getSpeedVector(x, double=False):
    m = len(x)
    if len(x.shape) < 2:
        x = np.expand_dims(x,1)
    d = np.sqrt(((x[1:]-x[:-1])**2).sum(axis=1))
    z = []
    zsum = 0
    for i in range(len(x)):
        z.append(zsum)
        if i < m-1:
            if double:
                zsum += d[i]/2
                z.append(zsum)
                zsum += d[i]/2
            else:
                zsum += d[i]
    return np.array(z)

def generateUniformKnotVector(n,k):
    t = np.linspace(0,1,n+2)
    t = np.concatenate(([0]*k, t, [1]*k)) #n+2k+2
    return t

def generateDistKnotVector(x, n, k):
    z = getSpeedVector(x)
    z /= z[-1]
    z = z[1:-1]
    t = []
    for i in range(n):
        t.append(z[round((len(z)-1)*(i+1)/(n+1))])
        
    t = np.concatenate(([0]*(k+1), t, [1]*(k+1)))
    return t

def generateKnot(knotStrategy,x,n,k):
    if knotStrategy == 0:
        t = generateDistKnotVector(x,n,k)
    elif knotStrategy == 1:
        t = generateUniformKnotVector(n,k)
    return t
    
def vieta_formula(roots, n):
    # Initialize an array for polynomial coefficients
    coeff = [0] * (n + 1)
    # Set the highest order coefficient as 1
    coeff[0] = 1
 
    for i in range(n):
        for j in range(i + 1, 0, -1):
            coeff[j] += roots[i] * coeff[j - 1]
            
    return coeff[::-1]
    
def constructSmoothingMatrix(t,n,k):
    AT = np.zeros((n+k+1,n))
    for j in range(n):
        for i in range(k+n+1):
            if i<j+1 or i>j+k+1:
                AT[i][j] = 0
            else:
                roots = np.array(t[i:i+k+2])
                roots = -roots
                # Call the vieta_formula function
                coeffs = vieta_formula(roots, len(roots))
                pi_prime = 0
                for ri in range(1,len(roots)+1):
                    pi_prime += coeffs[ri]*(t[j+k+1]**(ri-1))*ri
                
                #print(i,j, i+1, i-k, len(t), t[j], pi_prime)
                #print(math.factorial(k),(t[i+k+1]-t[i]),pi_prime, t[j+k+1], roots, coeffs)
                AT[i][j] = math.factorial(k)*(t[i+k+1]-t[i])/pi_prime
    return AT.T

def constructSecondDerMatrix(x):
    m = len(x)
    h = x[1:]-x[:-1]
    Q = np.zeros((m,m-2))
    R = np.zeros((m-2,m-2))
    for i in range(m-2):
        Q[i][i] = 1/h[i]
        Q[i+1][i] = -(1/h[i]+1/h[i+1])
        Q[i+2][i] = 1/h[i+1]
        R[i][i] = (h[i]+h[i+1])/3
        if i < m-3:
            R[i+1][i] = R[i][i+1] = h[i+1]/2
            
    return Q @ np.linalg.inv(R) @ Q.T
```

```python
def solveControlLS_old(x, y, w, t, k, lm, clamp):
    assert lm>=0 and lm<1
    x_norm = np.array(x)
    bs = bspline.Bspline(t, k)
    B = np.array([bs._Bspline__basis(xi,k) for xi in x_norm])
    B[-1][-1] = 1
    A = constructSmoothingMatrix(t,n,k)
    Bprime = B * np.repeat(np.expand_dims(w,1),n+k+1,axis=1)
    yprime = w*y
    if clamp:
        P = np.zeros((n+k+1,n+k+1))
        P[0][0] = P[-1][-1] = 1
        LHS = np.concatenate((2*(Bprime.T@Bprime)+((lm/(1-lm))*(A.T@A)), P.T), axis=1)
        LHS = np.concatenate((LHS, np.concatenate((P, np.zeros((n+k+1,n+k+1))), axis=1)), axis=0)
        Q = np.zeros(n+k+1)
        Q[0], Q[-1] = y[0], y[-1]
        RHS = np.concatenate((2*(Bprime.T @ yprime), Q))
        CZ = np.linalg.pinv(LHS) @ RHS
        C = CZ[:n+k+1]
    else:
        #C = np.linalg.pinv(Bprime) @ (yprime)
        C = np.linalg.pinv(Bprime.T @ Bprime + (lm/(1-lm)* (A.T@A))) @ Bprime.T @ yprime

    return C, 0#np.linalg.norm(yprime - (Bprime@C))**2, np.linalg.norm(A@C)**2
```

```python
def solveControlLSS(x, y, w, t, k, lm, clamp):
    assert lm>=0 and lm<1
    x = np.array(x)
    bs = bspline.Bspline(t, k)
    B = np.array([bs._Bspline__basis(xi,k) for xi in x])
    B[-1][-1] = 1
    A = constructSmoothingMatrix(t,n,k)
    #K = constructSecondDerMatrix(x)
    Bprime = B * np.repeat(np.expand_dims(w,1),n+k+1,axis=1)
    if clamp:
        Btilde = Bprime[1:-1,1:-1]
        Atilde = A[:,1:-1]
        s,rtilde = [],[]
        for j in range(1,m-1):
            rtilde.append(y[0]*B[j][0]+y[-1]*B[j][-1])
        for j in range(n):
            s.append(y[0]*A[j][0]+y[-1]*A[j][-1])
        rtilde, s = np.array(rtilde), np.array(s)
        ytilde = y[1:-1] - rtilde
        wtilde = w[1:-1]
        ytilde *= wtilde
        ctilde = np.linalg.pinv(Btilde.T @ Btilde + (lm/(1-lm)* (Atilde.T@Atilde))) @ (Btilde.T @ ytilde + ((lm/(1-lm)*Atilde.T@s)))
        C = np.concatenate(([y[0]], ctilde, [y[-1]]))
        obj1 = np.linalg.norm(ytilde-(Btilde @ ctilde))**2
        obj2 = np.linalg.norm(s+(Atilde @ ctilde))**2
    else:
        yprime = y*w
        C = np.linalg.pinv(Bprime.T @ Bprime + (lm/(1-lm)* (A.T@A))) @ Bprime.T @ yprime
        obj1 = np.linalg.norm(yprime - (Bprime@C))**2
        obj2 = np.linalg.norm(A@C)**2

    return C, 0#obj1, obj2
```

```python
def solveControlLS(x, y, w, t, k, lm, clamp):
    assert lm>=0 and lm<1
    x = np.array(x)
    m = len(x)
    bs = bspline.Bspline(t, k)
    B = np.array([bs._Bspline__basis(xi,k) for xi in x])
    B[-1][-1] = 1
    Btilde = B[1:-1,1:-1]
    #A = constructSmoothingMatrix(t,n,k)
    K = constructSecondDerMatrix(x)
    
    if clamp:
        ytilde = y[1:-1]
        Ktilde = K[1:-1,1:-1]
        stilde = y[0]*K[0,1:-1]+y[-1]*K[-1,1:-1]
        rtilde = y[0]*B[1:-1,0]+y[-1]*B[1:-1,-1]
        c = np.linalg.inv(2*lm*(Btilde.T @ Ktilde @ Btilde)+(2*(1-lm)*(Btilde.T@Btilde))) @ (-lm*(Btilde.T @ (K[1:-1,0]*y[0]+K[1:-1,-1]*y[-1]+stilde))+(2*(1-lm)*(Btilde.T @ (ytilde-rtilde))))
        c = np.concatenate(([y[0]],c,[y[-1]]))
        
    else:
        c = np.linalg.inv((1-lm)*(B.T@B)+lm*(B.T@K@B)) @ ((1-lm)*B.T@y)

    return c
```

```python
def solveControlLS(x, y, w, t, k, lm, clamp):
    assert lm>=0 and lm<1
    x = np.array(x)
    m = len(x)
    bs = bspline.Bspline(t, k)
    B = np.array([bs._Bspline__basis(xi,k) for xi in x])
    B[-1][-1] = 1
    K = constructSecondDerMatrix(x)
    #print("k")
    #print(K)
     
    if clamp:
        P = np.zeros((n+k+1,n+k+1))
        P[0][0] = P[-1][-1] = 1
        q = np.zeros(n+k+1)
        q[0], q[-1] = y[0], y[-1]
        LHS = np.concatenate(((1-lm)*B.T@B+lm*B.T@K@B, P.T), axis=1)
        LHS = np.concatenate((LHS, np.concatenate((P, np.zeros((n+k+1,n+k+1))), axis=1)), axis=0)
        RHS = np.concatenate(((1-lm)*(B.T @ y), q))
        cz = np.linalg.pinv(LHS) @ RHS
        c = cz[:n+k+1]
        
    else:
        c = np.linalg.inv((1-lm)*(B.T@B)+lm*(B.T@K@B)) @ ((1-lm)*B.T@y)

    return c
```

```python
def getTck2D(path, z, w, k, n, lm, clamp, knotStrategy):
    m=len(path)
    assert n>=0 and n<=m-k-1
    x,y = path[:,0], path[:,1]
    x,y = np.array(x), np.array(y)
    
    #generate knot vector t
    t = generateKnot(knotStrategy,path,n,k)
    t = t*(z.max()-z.min())+z.min()

    #Compute cx, cy
    cx = solveControlLS(z,x,w,t,k,lm=lm, clamp=clamp)
    cy = solveControlLS(z,y,w,t,k,lm=lm, clamp=clamp)
    c = np.array([cx,cy])
    tck = (t, c, k)
    return tck
```

```python
def gdk(z, n, k):
    zt = z[1:-1]
    t = []
    for i in range(n):
        t.append(zt[round((len(zt)-1)*(i+1)/(n+1))])
        
    t = np.concatenate(([z.min()]*(k+1), t, [z.max()]*(k+1)))
    return t

def getInterpolatingMidwayControls(path,z,k,cn=-1,lm=0,clamp=True):
    assert cn<=len(path)
    if cn<0:
        cn = len(path)
        
    t = gdk(getSpeedVector(path, double=True),cn*2-k-2,k)
    bs = bspline.Bspline(t, k)
    B = np.array([bs._Bspline__basis(zi,k) for zi in z])
    B[-1][-1] = 1
    tB = np.zeros((cn*2-1, cn))
    tB[0:2,0] = [1,0.5]
    tB[-2:,-1] = [0.5,1]
    for i in range(1,cn-1):
        tB[i*2-1:i*2+2,i] = [0.5,1,0.5]

    Bhat = B@tB
    K = constructSecondDerMatrix(z)
    if clamp:
        P = np.zeros((cn,cn))
        P[0][0] = P[-1][-1] = 1
        q = np.zeros((cn,path.shape[1]))
        q[0] = path[0]
        q[-1] = path[-1]
        RHS = np.concatenate(((1-lm)*Bhat.T@Bhat+lm*(Bhat.T@K@Bhat),P.T), axis=1)
        RHS = np.concatenate((RHS, np.concatenate((P,np.zeros((cn,cn))),axis=1)),axis=0)
        LHS = np.concatenate(((1-lm)*Bhat.T@path, q), axis=0)
        cz = np.linalg.lstsq(RHS,LHS,rcond=-1)[0]
        c = cz[:cn]
    else:
        c = np.linalg.solve((1-lm)*Bhat.T@Bhat+lm*(Bhat.T@K@Bhat), (1-lm)*Bhat.T@path)
        
    '''cn = np.zeros((len(c)*2-1,path.shape[1]))
    cn[::2] = c
    cn[1::2] = (c[1:] + c[:-1])/2
    cn = cn.T'''

    cn = (tB @ c).T
    
    return (t, cn, k)
```

# Obstacles

```python
def area_triangle_d(p1, p2, p3):
    return abs((p1[0]*(p2[1]-p3[1]) + p2[0]*(p3[1]-p1[1]) + p3[0]*(p1[1]-p2[1])))

def point_in_triangle(point, triangle,eps):
    triangle = np.array(triangle).reshape(3,2)
    v1, v2, v3 = triangle
    full_area = area_triangle_d(v1, v2, v3)
    sub_area1 = area_triangle_d(point, v2, v3)
    sub_area2 = area_triangle_d(v1, point, v3)
    sub_area3 = area_triangle_d(v1, v2, point)
    return np.abs((sub_area1 + sub_area2 + sub_area3) - full_area) - eps  # Use epsilon for floating point comparison
    
def barycentric_coordinates(x, y, vertices):
    x1, y1 = vertices[0]
    x2, y2 = vertices[1]
    x3, y3 = vertices[2]
    
    denominator = (y2 - y3)*(x1 - x3) + (x3 - x2)*(y1 - y3)
    alpha = ((y2 - y3)*(x - x3) + (x3 - x2)*(y - y3)) / denominator
    beta = ((y3 - y1)*(x - x3) + (x1 - x3)*(y - y3)) / denominator
    gamma = 1 - alpha - beta
    
    return alpha, beta, gamma

def convex_hull_cost(p, vertices, k=1):
    vertices = np.array(vertices).reshape(3,2)
    x,y = p
    alpha, beta, gamma = barycentric_coordinates(x, y, vertices)
    return k * alpha * beta * gamma
```

```python
import numpy as np
from scipy.optimize import minimize

# Define A, b, and the initial guess for x
P=[[3,2],
   [3.2,4.05],
   [4.25,4.5]]
P=np.array(P)
x0 = (np.linalg.pinv(Beq)@target).flatten()

# Define the objective function (squared norm)
def objective(x):
    return np.linalg.norm(Beq @ x.reshape(m,2) - target)

# Define the constraint function
def constraint(x,i,p):
    #return point_in_triangle(p,x[i*2:i*2+6],eps)
    return convex_hull_cost(p, x[i*2:i*2+6], k=1)

def hard_constraint(x,i,p):
    return max(0.0,convex_hull_cost(p, x[i*2:i*2+6], k=1))

# Define the constraint bounds
constraint_bounds = []#[{'type': 'ineq', 'fun': constraint, 'args':(i,p)} for i in range(2) for p in P]

# Perform the optimization
cl = 5000
penalty = lambda x: sum([hard_constraint(x,i,p) for i in range(6) for p in P])
obj = lambda x: np.linalg.norm(Beq @ x.reshape(m,2) - target) + cl*penalty(x)
result = minimize(obj, x0, method='CG', constraints=constraint_bounds, options={'maxiter':10000, 'eps':1e-15, 'gtol':1e-06}, tol=1e-04)

# Print the result
print("Optimal x:", result.x)
print("Minimum value of the objective function:", result.fun)
```
